### Exportando la data:

```python
# Ubicacion del archivo. Se antepone un ./ a la ruta para indicar que
# es una ruta relativa (osea, que usamos de origen la carpeta actual)

ruta_datos_exoplanetas = './oec.csv'

# pd: referencia a la libreria pandas
# .read_csv: acceso a la funcion de lectura de archivos '.csv'
exoplanets_df = pd.read_csv(ruta_datos_exoplanetas)

# Previsualizando el dataframe cargado
exoplanets_df
```

```python
exoplanets_df.columns
```

Estos son los atributos que usaremos: 
* PlanetIdentifier: Es el nombre identificador de planetas.
* PlanetaryMassJpt: Es la masa del planeta como múltiplo de la masa de Júpiter. 
* RadiusJpt: La medida del radio como múltiplo del radio de Júpiter.  
* SurfaceTempK: Es la temperatura en Kelvin de la superficie del planeta.

```python
# Atributos utiles para nuestros fines
useful_features = ['PlanetIdentifier', 'PlanetaryMassJpt', 'RadiusJpt', 'SurfaceTempK']

# Recortando el dataframe original para usar solo lo estrictamente necesario
# exoplanets_df: Nuestro dataframe hasta ahorita
# .loc: Se puede entender como "localizar"
# ¿Localizar qué exactamente?: Revisando lo que se ve entre corchetes:
# [<aquí se localizan filas> , <aquí se localizan las columnas>]
# Para nuestro caso [: , useuseful_features] considerará TODAS las filas (por usar dos puntos ":")
# y las columnas provistas en la lista "useful_features" recien creada.
# El resultado de este recorte usando ".loc" se guardará en el mismo dataframe "exoplanetes_df"
# sobreescribiendo su estructura original

exoplanets_df = exoplanets_df.loc[:, useful_features]
exoplanets_df
```

Una vez fijados los atributos con los que trabajaremos a continuación se procederá a limpiar el dataframe de los valores no numéricos (NaN) que estorbarían el proceso analítico.  

```python
# Tomamos el dataframe en su estado actual (ya recortado por el paso anterior)
# y "dropeamos" (descartamos/tiramos) los valor no numericos

exoplanets_df = exoplanets_df.dropna()
exoplanets_df
```

***
Para cumplir con las unidades del Sistema Internacional cambiaremos el dataframe original que está en múltiplos de Jupiter por valores normales como la masa en kilogramos (Kg) y el radio en metros (m).

```python
jupyter_mass = 1.898e27 # kg
jupyter_radius = 69911000 # m

exoplanets_df['PlanetaryMass_kg'] = exoplanets_df['PlanetaryMassJpt']*jupyter_mass
exoplanets_df['Radius_m'] = exoplanets_df['RadiusJpt']*jupyter_radius

exoplanets_df
```

***
Para poder analizar que planetas cumplen con el rango habitable de un ser humano se considerará la gravedad como factor para la supervivencia. 
Dado que la gravedad no se encuentra como atributo originalmente, se creará una columna donde se calculará su valor para cada planeta. 

Se usará la fórmula de gravitación universal:

$$ g_{planeta} =  G. \frac{M_{planeta}}{R_{planeta}^2} $$

* $g_{planeta}$= Gravedad del planeta.
* $G$= Constante de gravitación universal (G=$6.674x10^{-11}N.m^{2}/Kg^{2}$
* $M_{planeta}$=Masa del planeta. 
* $R_{planeta}$=Radio del planeta.

```python
# Creando nueva columna llamada 'gPlanet'
G = 6.67408e-11

# Si bien "exoplanets_df['PlanetaryMassJpt']" es una columna (siendo mas exigentes con
# la terminología, es una Serie de pandas) python nos permite operar
# directamente haciendo cosas como aplicar la formula de gravitacion como se muestra a continuacion:

# Aplicando gravitacion universal
exoplanets_df['gPlanet'] = G * exoplanets_df['PlanetaryMass_kg']/(exoplanets_df['Radius_m']**2)
exoplanets_df['gPlanet'] = exoplanets_df['gPlanet']/9.81

exoplanets_df
```

___
En este link encontramos datos sobre la resistencia de un ser humano ante la gravedad:
 https://www.ambientum.com/ambientum/ciencia/gravedad-maxima-soportar-humanos.asp este estudio se basa en este otro artículo:
 https://arxiv.org/pdf/1808.07417.pdf .
 
 En resumen: 
 Un ser humano podría resistir:
 * En los huesos hasta 10 veces la gravedad de la tierra, mayor a eso los huesos se partirían en mil pedazos. 
 * En los músculos y órganos hasta 5 veces la gravedad de la tierra, mayor a eso un ser humano no podría siquiera moverse.
 * En la sangre hasta 4 veces la gravedad de la tierra, dado que la sangre tiende a bajar a las piernas y el corazón requiere de      mayor esfuerzo para bombear la sangre hasta el cerebro. 
 
 De todo esto se concluye que lo máximo que soporta un ser humano es 4 veces la gravedad de la tierra, mayor a esto se moriría. 

___

Este link menciona el rango de supervivencia de un ser humano ante la temperatura: https://www.microsiervos.com/archivo/ciencia/calor-extremo-humano-morir-60-c.html . 

En resumen un ser humano a los:
* 37°: Temperatura normal del cuerpo humano.
* 42°: Deshidratación.
* 50°: Temperatura récord hasta ahora (India).
* 55°: Temperatura máxima que soporta un ser humano.
* 60°: Muerte a los 10 minutos.
* 70°: Muerte inmmediata. 

Entonces se afirma que 60°C es lo máximo que soporta el cuerpo humano.
